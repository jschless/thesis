# Predicting Price Using a Linear Model

```python
%matplotlib inline

import sklearn
import pandas as pd
from sklearn import datasets
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
boston = datasets.load_boston()
digits = datasets.load_digits()


boston.feature_names

bos_pd = pd.DataFrame(boston.data, columns=boston.feature_names)
bos_pd['PRICE'] = boston.target

Y = bos_pd['PRICE']
X = bos_pd.drop('PRICE', axis = 1)
#X = bos_pd.filter(['CRIM'])
lm = LinearRegression()
gamma = .8
train_size = int(gamma * len(Y))
trainY, testY = Y[0:train_size], Y[train_size:]
trainX, testX = X[0:train_size], X[train_size:]
lm.fit(trainX, trainY)
predictY = lm.predict(testX)
linScore = lm.score(testX, testY)
print("Linear regression score: " + str(linScore))
plt.scatter(predictY, testY)
plt.title("Predictions v Reality")
plt.ylabel("Actual Cost")
plt.xlabel("Predicted Cost")
bos_pd.head()
```

# Experimenting with Ridge

* Ridge regression tries to rectify problems with using least-squares regression on multicolinear data (independent variables have linear relationships). Least-squares has large variances in this case, and ridge regression reduces standard errors.

#### Steps to Ridge Regression
1. Standardize variables.
2. $\hat{B} = (R + kI)^{-1}X'Y $ 
3. Adding $k$ to diagonal elements of the correlation matrix. For some value $k$, mean squared error for ridge estimation is less than least squares.


```python
from sklearn.linear_model import Ridge
x = int(.8*len(trainX))
tSetX, tSetY = trainX[0:x], trainY[0:x]
vSetX, vSetY = trainX[x:], trainY[x:]
best, val = -1000, -1000
alphas = [.1, .5, .8, 1.0, 1.5, 2.0, 10.0]
for a in alphas:
    clf = Ridge(alpha=a)
    clf.fit(tSetX, tSetY)
    score = clf.score(vSetX, vSetY)
    print("score for alpha= " + str(a) + " is " + str(score))
    if score > best:
        val = a
    plt.figure()
    plt.title("Ridge with alpha=" + str(a))
    plt.bar(x=boston.feature_names, height=clf.coef_)
clf = Ridge(alpha = val)
clf.fit(trainX, trainY)
score = clf.score(testX, testY)
print("overall score with alpha=" + str(val) + " is " + str(score))
```

# Experimenting with Lasso
* performs variable selection, so probably will be better than brute force approaches above
* Uses l1 norm



```python
from sklearn.linear_model import Lasso

x = int(.8*len(trainX))
tSetX, tSetY = trainX[0:x], trainY[0:x]
vSetX, vSetY = trainX[x:], trainY[x:]
best, val = -1000, -1000
alphas = [.1, .5, .8, 1.0, 1.5, 2.0, 10.0]
for a in alphas:
    clf = Lasso(alpha=a)
    clf.fit(tSetX, tSetY)
    score = clf.score(vSetX, vSetY)
    print("score for alpha= " + str(a) + " is " + str(score))
    if score > best:
        val = a
    plt.figure()
    plt.title("Lasso with alpha=" + str(a))
    plt.ylabel("Weight")
    plt.bar(x=boston.feature_names, height=clf.coef_)
clf = Lasso(alpha = val)
clf.fit(trainX, trainY)
score = clf.score(testX, testY)
print("overall score with alpha=" + str(val) + " is " + str(score))
print(clf.coef_)
```

### Observations
* As alpha increase, one weight starts to dominate the rest of the variables. 
* With Lasso, as alpha increases fewer variables are selected.
* With Ridge, an alpha increase lowers the coefficients for all but one variable, which increases.

# K-Fold Cross Validation

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5)

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = Y.iloc[train_index], Y.iloc[test_index]
    alphas = [.1, .5, .8, 1.0, 1.5, 2.0, 10.0]
    for a in alphas:
        clf = Lasso(alpha=a)
        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)
        print("score for alpha= " + str(a) + " is " + str(score))
```

# Notes on Classification
### Effects of C
* Regularization prevents overfitting. We don't want our coefficients to get too high because we want a flexible and generalizeable model. So, by adding some $\lambda \cdot \sum_{i=1} \beta_i^2 $ to our error function, we can try to control overfitting because the model seeks to minimize the error function. By adding this term, the error function is concerned with the size of the coefficients in addition to the size of the error. 
* This tuning parameter is $C$ or $\alpha$ in the above models
* $C$ is the inverse. As such, a small $C$ represents very high regularization. This explains why when $C=.001$, every coefficient was 0. The penalty was simply too high to make a coefficient nonzero. That's why the visualization appears to be completely black.

### l1 v l2 penalties
* The l2 regulatization adds squared magnitude of coefficients. $\lambda \cdot \sum_{i=1} \beta_i^2 $
* The l1 regularization adds absolute value of magnitude. $\lambda \cdot \sum_{i=1} |\beta | $
* l2 is ridge, l2 is lasso. Variations of a normal linear regression.
